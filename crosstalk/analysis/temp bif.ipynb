```python
def plotDF(df1,xkey):
    label={'A':'AMPK','h':'Hif-1'}
    clist=np.array(['r','k','b'])
    fig = plt.figure()
    gs1 = gridspec.GridSpec(2, 3)
    gs1.update(left=0.05, right=2, wspace=0.5,hspace=0.3,top=2,bottom=0)

    axC1 = plt.subplot(gs1[0, 0])
    axC2 = plt.subplot(gs1[0, 1])
    axC3 = plt.subplot(gs1[0, 2])
    axC4 = plt.subplot(gs1[1, 0])
    axC5 = plt.subplot(gs1[1, 1])
    #axC6 = plt.subplot(gs1[1, 2])
    
    color_arr = 1*(df1['stability'].values=='S')+2*(df1['stability'].values=='U')
    axC1.scatter(df1[xkey],df1['u'],c=clist[color_arr])
    axC2.scatter(df1[xkey],df1['u3'],c=clist[color_arr])
    axC3.scatter(df1[xkey],df1['mz'],c=clist[color_arr])
    axC4.scatter(df1[xkey],df1['ms'],c=clist[color_arr])
    axC5.scatter(df1[xkey],df1['S'],c=clist[color_arr])
    
    axC1.set_xlabel(label[xkey])
    axC2.set_xlabel(label[xkey])
    axC3.set_xlabel(label[xkey])
    axC4.set_xlabel(label[xkey])
    axC5.set_xlabel(label[xkey])

    axC1.set_ylabel('$\mu_{200}$')
    axC2.set_ylabel('$\mu_{34}$')
    axC3.set_ylabel('Zeb mRNA')
    axC4.set_ylabel('Snail mRNA')
    axC5.set_ylabel('Snail')
    #axC6.scatter(df1[xkey],df1['h'],c=clist[1*(df1['stability'].values=='S')])
    #axC6.set_xlabel(label[xkey])
    #axC6.set_ylabel('Hif-1')
```

```python
df1 = pd.read_csv('../bifurcations/data_bif/EMW_bif_lamdahu_0.85.txt',header=0)
plotDF(df1,'h')
```
