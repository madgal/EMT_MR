```python
def get(lvx,lvy,xk=[],yk=[],count=0,tmpx=[],tmpy=[],finx=[],finy=[]):

    if xk==[] and yk==[]:
        xk = lvx.keys()
        yk = lvy.keys()
    if count==(len(xk)+len(yk)-1):
        ntx,nty = [],[]
        for i in range(len(lvy[yk[-1]])):
            ty = tmpy+[lvy[yk[-1]][i]]
            ntx+=[tmpx]
            nty+=[ty]
        finx+=ntx
        finy+=nty
        return finx,finy
    else:
        if count <len(xk):
            for i in range(len(lvx[xk[count]])):
                t2 = tmpx+[lvx[xk[count]][i]]
                finx,finy=get(lvx,lvy,xk,yk,count+1,t2,tmpy,finx,finy)
        elif count <len(xk)+len(yk):
            for i in range(len(lvy[yk[count%len(xk)]])):
                t2 = tmpy+[lvy[yk[count%len(xk)]][i]]
                finx,finy=get(lvx,lvy,xk,yk,count+1,tmpx,t2,finx,finy)
        else:
            return finx,finy
        
    return finx,finy
        
        
```
