```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plotNull(xkey,ykey,fpTitle,end):
    if xkey=='A':
        f1 = 'data/nullA_'+end+".txt"
    else:
        f1 = 'data/nullU_'+end+".txt"
    if ykey=='h':
        f2 = 'data/nullh_'+end+".txt"
    else:
        f2 = 'data/nullMZ_'+end+".txt"

    df_fp = pd.read_csv("data/"+fpTitle)
    df_null1 = pd.read_csv(f1)
    df_null2 = pd.read_csv(f2)
    print df_fp

    plt.plot(df_null1[xkey],df_null1[ykey],'g')
    plt.plot(df_null2[xkey],df_null2[ykey],'b')
    plt.plot(df_fp[xkey],df_fp[ykey],'ok')
    plt.show()
```

```python
plotNull('A','h','MR_fp.txt','mr')
```
